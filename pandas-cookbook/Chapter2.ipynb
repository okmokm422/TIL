## カラムを意味がわかるように順序付け

- 連続か離散か？

- グループ分けできないか？

- 最も重要なカラム郡を前に

```python
movie = pd.read_csv('data/movie.csv')
movie.head()
```

```python
movie.columns
```

```python
disc_core = ['movie_title', 'title_year', 
             'content_rating', 'genres']
disc_people = ['director_name', 'actor_1_name', 
               'actor_2_name', 'actor_3_name']
disc_other = ['color', 'country', 'language', 
              'plot_keywords', 'movie_imdb_link']
cont_fb = ['director_facebook_likes', 'actor_1_facebook_likes', 
           'actor_2_facebook_likes', 'actor_3_facebook_likes', 
          'cast_total_facebook_likes', 'movie_facebook_likes']
cont_finance = ['budget', 'gross']
cont_num_reviews = ['num_voted_users', 'num_user_for_reviews',
                   'num_critic_for_reviews']
cont_other = ['imdb_score', 'duration', 
              'aspect_ratio', 'facenumber_in_poster']
```

```python
# リストを順番にまとめて連結
new_col_order = disc_core + disc_people + \
disc_other + cont_fb + \
cont_finance + cont_num_reviews + \
cont_other
```

```python
# 元のカラムが全てリストに含まれているか確認
set(movie.columns) == set(new_col_order)
```

```python
# カラム順序を新しくしたリストをDataFrameのインデックス演算子に
movie2 = movie[new_col_order]
movie2.head()
```

## DataFrameの演算子の動き

```python
college = pd.read_csv('data/college.csv')
college.head()
```

```python
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
college_ugds_ + .00501
```

```python
(college_ugds_ + .00501) // .01
```

```python
college_ugds_op_round   = (college_ugds_ + .00501) // .01 / 100
college_ugds_op_round.head()
```

- [round(number[, ndigits])](https://docs.python.org/ja/3/library/functions.html#round)

```python
round(0.5)
```

```python
round(0.5+0.1)
```

```python
round(0+0.1)
```

```python
# 真ん中の数（0.5）は偶数に丸まってしまうので、丸めの前に端数を加えておく
college_ugds_round = (college_ugds_ + .00001).round(2)
college_ugds_round.head()
```

```python
# 丸め方が違っても数値は同じかを確認
college_ugds_op_round.equals(college_ugds_round)
```

```python
# pythonの浮動小数点の不正確さ
.045 + .005
```

```python
# 同じ計算をメソッドで
college_ugds_op_round_methods = college_ugds_.add(.00501).floordiv(.01).div(100)
```

```python
college_ugds_op_round_methods.equals(college_ugds_op_round)
```

## 欠損値の比較

- pandasは欠損値をnumpyのNaNオブジェクトで返す。（np.nan）

- これは自分自身とも等しくない変わったオブジェクト。

```python
# 等しくない
import numpy as np
np.nan == np.nan
```

```python
None == None
```

```python
# 不等価演算子を除いて比較演算子にはFalseを返す
np.nan < 5
```

```python
# 不等価演算子を除いて比較演算子にはFalseを返す
5 > np.nan
```

```python
# 不等価演算子を除いて比較演算子にはFalseを返す
np.nan != 5
```

```python
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
```

```python
college_ugds_ == .0019
```

```python
# 自分自身との比較（indexラベルの比較／要素の個数が同じかどうか）
college_self_compare = college_ugds_ == college_ugds_
college_self_compare.head()
```

```python
# カラムにTrueしか含まれないかをall()メソッド（全て同じ値か）で確認
# 結果がFalseになるのは、欠損値同士の比較がFalseになってしまうため。
college_self_compare.all()
```

```python
# np.nan==np.nanはFalseなので合計値がゼロになる
(college_ugds_ == np.nan).sum()
```

```python
# 欠損値を調べるにはisnull()で
college_ugds_.isnull().sum()
```

```python
# equalsメソッドで比較する
college_ugds_.equals(college_ugds_)
```

```python
# college_ugds_ == .0019と同じ
college_ugds_.eq(.0019)
```

```python
# ２つのデータフレームが同じかを確認
# 等しいとNoneを返す
from pandas.testing import assert_frame_equal
assert_frame_equal(college_ugds_,college_ugds_)
```

## 大学のキャンパスのダイバーシティ指標の計算

```python
pd.read_csv('data/college_diversity.csv',index_col='School')
```

```python
college = pd.read_csv('data/college.csv',index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
```

```python
# 欠損値のある大学を確認 isnullをsumしてdescendingに
college_ugds_.isnull().sum(axis=1).sort_values(ascending=False).head()
```

```python
# 欠損値を削除した上で合計
# how ='all'にすることで全て欠損値の行を削除
college_ugds_ = college_ugds_.dropna(how='all')
college_ugds_.isnull().sum()
```

- [DataFrame.ge(self, other, axis='columns', level=None)](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ge.html)

```python
# 欠損値を削除したデータが、0.15と等しい大きいか
college_ugds_.ge(.15)
```

```python
# 人種カテゴリが0.15以上=>Trueの個数を数える
diversity_metric = college_ugds_.ge(.15).sum(axis='columns')
diversity_metric
```

```python
# unique数を数えるvalue_counts
diversity_metric.value_counts()
```

```python
# 人種カテゴリが0.15以上の人種が5つの大学を探す
# sort_valueを降順に
diversity_metric.sort_values(ascending=False)
```

```python
college_ugds_.loc[['Regency Beauty Institute-Austin','Central Texas Beauty College-Temple']]
```

```python
diversity_metric.loc[us_news_top]
```

```python
# 最大人種の%が最大=> 多様性なし を調べる方法
college_ugds_.max(axis=1).sort_values(ascending=False).head(10)
```
