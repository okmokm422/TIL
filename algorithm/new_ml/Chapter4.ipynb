### 4.7.1 指数

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-4, 4, 100)
y = 2 ** x
y2 = 3 ** x
y3 = 0.5 ** x
```

```python
plt.figure(figsize=(5,5))

plt.plot(x, y, 'black', linewidth=3, label='$y=2^x$')
plt.plot(x, y2, 'cornflowerblue', linewidth=3, label='$y=3^x$')
plt.plot(x, y3, 'gray', linewidth=3, label='$y=0.5^x$')

plt.ylim(-2, 6)
plt.xlim(-4, 4)

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

### 4.7.2 対数

```python
x = np.linspace(-8,8,100)
y = 2**x

x2 = np.linspace(0.001, 8, 100) # np.log(0)はエラーになるため、スタートを限りなく0に近く
y2 = np.log(x2)/np.log(2) # np.logは自然対数のため、底を2としたlogに変換

plt.plot(x, y, 'black', linewidth=3, label='$y=2^x$')
plt.plot(x2, y2, 'cornflowerblue', linewidth=3, label='$y=\log{2} x$')
plt.plot(x, x, 'gray',linestyle=':', linewidth=3)

plt.ylim(-8, 8)
plt.xlim(-8, 8)

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

#### 自然対数

##### logを取ることの利点
- 大きすぎる数や小さすぎる数を扱いやすくしてくれることでオーバーフローを防ぐ
- 掛け算を足し算に変換する

```python
x = np.linspace(-4, 4, 100)
y = (x - 1)**2 + 2
logy = np.log(y)

plt.figure(figsize=(4,4))

plt.plot(x,y,'black',linewidth=3,label='$y=(x-1)^2+2$')
plt.plot(x,logy,'cornflowerblue',linewidth=3, label='$y=\logy$')

# 刻み幅をnp.arange(start, stop, step)で指定
plt.yticks(range(-4,9,1)) 
plt.xticks(range(-4,5,1))

plt.ylim(-4,8)
plt.ylim(-4,8)

plt.legend()
plt.grid(True)
plt.show()
```
