### 4.7.1 指数

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-4, 4, 100)
y = 2 ** x
y2 = 3 ** x
y3 = 0.5 ** x
```

```python
plt.figure(figsize=(5,5))

plt.plot(x, y, 'black', linewidth=3, label='$y=2^x$')
plt.plot(x, y2, 'cornflowerblue', linewidth=3, label='$y=3^x$')
plt.plot(x, y3, 'gray', linewidth=3, label='$y=0.5^x$')

plt.ylim(-2, 6)
plt.xlim(-4, 4)

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

### 4.7.2 対数

```python
x = np.linspace(-8,8,100)
y = 2**x

x2 = np.linspace(0.001, 8, 100) # np.log(0)はエラーになるため、スタートを限りなく0に近く
y2 = np.log(x2)/np.log(2) # np.logは自然対数のため、底を2としたlogに変換

plt.plot(x, y, 'black', linewidth=3, label='$y=2^x$')
plt.plot(x2, y2, 'cornflowerblue', linewidth=3, label='$y=\log_2 x$')
plt.plot(x, x, 'gray',linestyle=':', linewidth=3)

plt.ylim(-8, 8)
plt.xlim(-8, 8)

plt.grid(True)
plt.legend(loc='lower right')

plt.show()
```

#### 自然対数

##### logを取ることの利点
- 大きすぎる数や小さすぎる数を扱いやすくしてくれることでオーバーフローを防ぐ
- 掛け算を足し算に変換する

```python
x = np.linspace(-4, 4, 100)
y = (x - 1)**2 + 2
logy = np.log(y)

plt.figure(figsize=(4,4))

plt.plot(x,y,color='black',linewidth=3,label='$y=(x-1)^2+2$')
plt.plot(x,logy,color='cornflowerblue',linewidth=3, label='$y=\logy$')

# 刻み幅をnp.arange(start, stop, step)で指定
plt.yticks(range(-4,9,1)) 
plt.xticks(range(-4,5,1))

plt.ylim(-4,8)
plt.ylim(-4,8)

plt.legend()
plt.grid(True)
plt.show()
```

### 4.7.3 指数関数の微分

#### $a^x$の微分

```python
x = np.linspace(-4, 4, 100)
a = 2
y = a**x
dy = y * np.log(a)

plt.figure(figsize=(4,4))
plt.plot(x,y,'gray',linestyle='--',linewidth=3,label='$y=a^x$')
plt.plot(x,dy,'cornflowerblue',linewidth=3,label='$y=a^xloga$')

plt.ylim(-1,8)
plt.xlim(-4,4)
plt.grid(True)
plt.legend()
plt.show()
```

#### $e^x$の微分

```python
x = np.linspace(-4, 4, 100)
a = np.e
y = a**x
dy = y * np.log(a)

plt.figure(figsize=(4,4))
plt.plot(x,y,'gray',linestyle='--',linewidth=3,label='$y=e^x$')
plt.plot(x,dy,'cornflowerblue',linewidth=3,label='$y=e^xloga$')

plt.ylim(-1,8)
plt.xlim(-4,4)
plt.grid(True)
plt.legend()
plt.show()
```

### 4.7.4 対数関数の微分

```python
x = np.linspace(0.0001,4,100)
y = np.log(x)
dy = 1/x

plt.figure(figsize=(4,4))
plt.plot(x,y,'gray',linestyle='--',linewidth=3,label='$y=logx$')
plt.plot(x,dy,'cornflowerblue',linewidth=3,label='$y=\frac{1}{x}$')

plt.ylim(-8,8)
plt.xlim(-1,4)
plt.grid(True)
plt.legend()
plt.show()
```

### 4.7.6 ソフトマックス関数

$\displaystyle y_i=\frac{exp(x_i)}{\sum_{j=0}^{K-1} exp(x_j)}$

```python
# ３つの数値の大小関係を保ったまま確率に変換するソフトマックス関数
def softmax(x0, x1, x2):
    u = np.exp(x0) + np.exp(x1) + np.exp(x2)
    return np.exp(x0) / u, np.exp(x1) / u, np.exp(x2) / u
```

```python
# test
y = softmax(2, 1, -1)
print(np.round(y, 2)) # 小数点以下２桁を表示
print(np.sum(y)) # 和を表示
```

```python
# ソフトマックス関数を図示
from mpl_toolkits.mplot3d import Axes3D

# サンプル数
xn = 20
# 等差数列をつくる linspace(start, stop, num)
x0 = np.linspace(-4, 4, xn)
x1 = np.linspace(-4, 4, xn)
# zeros(num,rows,columns)でゼロ行列をつくる
y = np.zeros((xn, xn, 3))

# xn回分繰り返し
for i0 in range(xn):
    for i1 in range(xn):
        # ソフトマックス関数に数値を代入
        y[i1, i0, :] = softmax(x0[i0], x1[i1], 1)
```

```python
# (x0, x1)座標を返すmeshgrid
xx0, xx1 = np.meshgrid(x0, x1)
plt.figure(figsize=(10,5))

for i in range(2):
    # subplot(rows, cols)で図の出し方（数と配列）を決める
    ax = plt.subplot(1, 2, i+1, projection='3d')
    
    ax.plot_surface(xx0, xx1, y[:, :, i],
                   rstride=1, cstride=1, alpha=0.3,
                   color='blue', edgecolor='black')
    
    ax.set_xlabel('$x_0$', fontsize=14)
    ax.set_ylabel('$x_1$', fontsize=14)
    ax.view_init(40, -125)
```

### 4.7.8 ガウス関数

$y= exp(-x^2)$ 

$\displaystyle y = a\,exp \left(\cfrac{(x-\mu)^2}{2\sigma^2} \right)$

```python
# ガウス関数の定義
def gauss(mu, sigma, a):
          return a * np.exp(-(x - mu)**2 / (2 * sigma**2))
```

```python
x = np.linspace(-4, 4, 100)

plt.figure(figsize=(4, 4))
plt.plot(x, gauss(0, 1, 1), 'cornflowerblue', linewidth=3, label='μ=0,σ=1,a=1')
plt.plot(x, gauss(2, 2, 0.5), 'black', linewidth=3, label='μ=2,σ=2,a=0.5')

plt.ylim(-.5, 1.5)
plt.xlim(-4, 4)
plt.grid(True)
plt.legend()
plt.show()
```
