#### pandas_datareader

```python
# トヨタ自動車の銘柄コード
df = pandas_datareader.data.DataReader('7203.JP', 'stooq')
```

```python
df.head()
```

```python
plt.plot(df['Open'])
```

```python
df['Open'].plot()
```

```python
df['Open'].plot()
df['Open'].rolling(5).mean().plot() # 5日移動平均
df['Open'].rolling(25).mean().plot() # 25日移動平均
df['Open'].rolling(75).mean().plot() # 75日移動平均
```
