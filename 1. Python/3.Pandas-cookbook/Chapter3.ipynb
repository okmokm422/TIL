## 21. データ型を変更してメモリ削減

```python
# データの読み込み
college = pd.read_csv('data/college.csv')
```

```python
# カラムの選択
different_cols = ['RELAFFIL', 'SATMTMID', 'CURROPER', 'INSTNM', 'STABBR']
col2 = college.loc[:, different_cols]
col2.head()
```

```python
# 各カラムのデータ型を示す
col2.dtypes
```

- [DataFrame.memory_usage(self, index=True, deep=False)](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.memory_usage.html?highlight=memory_usage#pandas-dataframe-memory-usage)

```python
# memory_usageで各カラムのメモリ使用量を示す
original_mem = col2.memory_usage(deep=True)
original_mem
```

```python
# RELAFFILカラムは0/1なので、8ビット（1バイト）整数に変換
# astypeでデータ型の変換
col2['RELAFFIL'] = col2['RELAFFIL'].astype(np.int8)
```

```python
# 各カラムのデータ型を示す
col2.dtypes
```

```python
# memory_usageで各カラムのメモリ使用量を再度確認
college[different_cols].memory_usage(deep=True)
```

```python
# objectデータ型でデータの濃度（異なる値の個数）が低いものを探す
col2.select_dtypes(include=['object']).nunique()
```

- [note.nkmk.me / pandasでユニークな要素の個数、頻度（出現回数）をカウント](https://note.nkmk.me/python-pandas-value-counts/)

```python
# データの濃度が低いカラムをカテゴリ型に変換する
col2['STABBR'] = col2['STABBR'].astype('category')
col2.dtypes
```

```python
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
new_mem / original_mem
```

## 22. 最大の中の最小を選択

```python
# データの読み込み、columnを選んで新しいDataFrameを作る
movie = pd.read_csv('data/movie.csv')
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.head()
```

- [DataFrame.nlargest(self, n, columns, keep='first') ](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nlargest.html)  
    Return the first n rows ordered by columns in descending order.

```python
# nlargestメソッドを使い、imdb_scoreカラム上位100本の映画を見せる
movie2.nlargest(100, 'imdb_score').head()
```

- [DataFrame.nsmallest(self, n, columns, keep='first')](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nsmallest.html#pandas.DataFrame.nsmallest)  
    Return the first n rows ordered by columns in ascending order.

```python
# nsmallestメソッドをチェイニングして、上位100本の映画の中から最安値の5本の映画を返す
movie2.nlargest(100, 'imdb_score').nsmallest(5, 'budget')
```

## 25. トレール注文の価格計算

- [pandas-datareaderで株価や人口のデータを取得](https://note.nkmk.me/python-pandas-datareader-stock-population/)

```python
import pandas_datareader as pdr
```

```python
# startがうまく動かない？
tsla = pdr.DataReader('tsla',data_source='stooq',start='2020-3-1')
```

```python
tsla.index
```

```python
tsla = tsla.sort_index()
```

```python
tsla.head(8)
```

```python
# 取引日の終値を使う
tsla_close = tsla['Close']
tsla_close
```

- [Series.cummax(self, axis=None, skipna=True, *args, **kwargs)](https://pandas.pydata.org/docs/reference/api/pandas.Series.cummax.html)  
    Return cumulative maximum over a DataFrame or Series axis.
    

- [DataFrame.cummax(self, axis=None, skipna=True, *args, **kwargs)](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.cummax.html)

```python
# 過去からの最高値をトラック
tsla_cummax = tsla_close.cummax()
tsla_cummax.head(8)
```

```python
# トレール注文の価格（過去の最高値×0.9）
tsla_trailing_stop = tsla_cummax * .9
tsla_trailing_stop.head(8)
```

```python
# トレール注文の価格を返す関数
def set_trailing_loss(symbol, purchase_date, perc):
    close = pdr.DataReader(symbol, data_source='stooq', start=purchase_date)['Close']
    return close.cummax() * perc
```

```python
msft_trailing_stop = set_trailing_loss('msft', '2017-06-01', .85)
msft_trailing_stop.head()
```

```python
# 最小値をトラックするcomminメソッド
weight = pd.Series(data=[100,90,80,80,80,70,60,60,50])
pd.Series.cummin(weight) * 1.05
```
