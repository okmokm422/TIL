```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.offline
import plotly.express as px
import plotly.graph_objects as go
import plotly.subplots
import squarify
```

## 02 ヒストグラム：データのばらつき

```python
tips = sns.load_dataset("tips")
tips
```

```python
# ヒストグラム
sns.histplot(tips["total_bill"], kde=False)
```

```python
# ビンを増やす
sns.histplot(tips["total_bill"], kde=False, bins=15)
```

```python
sns.displot(y=tips["total_bill"], kde=False)
```

```python
# 2つのヒストグラムを描写する
lunch_tips = tips[tips["time"] == "Lunch"]
dinner_tips = tips[tips["time"] == "Dinner"]

sns.histplot(lunch_tips["total_bill"], bins=20, color="r")
sns.histplot(dinner_tips["total_bill"], bins=20)
```

## カウントプロット：量的変数の分布

```python
sns.countplot(x="smoker", data=tips)
```

```python
sns.countplot(x="smoker", hue="day", data=tips)
```

## 03 ボックスプロット：基本統計量を用いた分布

```python
sns.boxplot(y="total_bill", data=tips)
```

```python
sns.boxplot(x="time", y="total_bill", data=tips)
```

```python
# boxplotで引数orderを指定
sns.boxplot(x="time", y="total_bill", data=tips, order=["Dinner", "Lunch"])
```

```python
# 属性によって細かく分ける
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker")
```

## 04 散布図：２つのデータの関係性

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# 色を変える
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

```python
# 引数style
sns.scatterplot(x="total_bill", y="tip", data=tips, style="time")
```

```python
# 重なりが多いときの工夫
# alphaで点を透過させる
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time", alpha=0.5)
```
