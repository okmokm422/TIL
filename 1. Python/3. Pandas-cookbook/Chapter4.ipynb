# データから部分抽出

## 26 Seriesデータの選択

````python
import pandas as pd
import numpy as np
````

````python
college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
college.head()
````

````python
city = college['CITY']
city.head()
````

#### ilocインデクサ

````python
city.iloc[3]
````

````python
city.iloc[[10,20,30]]
````

````python
# start:stop:stepのスライス表記
city.iloc[4:50:10]
````

#### locインデクサ

````python
city.loc['Heritage Christian University']
````

````python
type(city.loc['Heritage Christian University'])
````

- [numpy.random.seed(self, seed=None)](https://numpy.org/devdocs/reference/random/generated/numpy.random.seed.html?highlight=random.seed#numpy.random.seed)

````python
# シード値を１に指定
np.random.seed(1)
````

- [numpy.random.choice(a, size=None, replace=True, p=None)](https://numpy.org/devdocs/reference/random/generated/numpy.random.choice.html?highlight=random.choice#numpy.random.choice)  
    Generates a random sample from a given 1-D array

````python
# 指定したシード値で大学名抽出
labels = list(np.random.choice(city.index, 4))
labels
````

````python
# リストでスライシング
city.loc[labels]
````

````python
# Seriesが返る
type(city.loc[labels])
````

````python
# 文字列でスライシング
city.loc['Alabama State University': 'Reid State Technical College': 10]
````

````python
# Seriesが返る
type(city.loc['Alabama State University': 'Reid State Technical College': 10])
````

````python
# Seriesで返したいときは、スカラー値ではなく、リストでindexを渡す。
city.iloc[[3]]
````

````python
# Seriesが返る
type(city.iloc[[3]])
````

````python
# 該当がないと空のSeriesが返る
city.loc['Reid State Technical College':'Alabama State University':10]
````

## 27 DataFrameの行の選択

````python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college.head()
````

````python
# indexに整数を渡して行選択
college.iloc[60]
````

````python
# インデックスラベルを渡して行選択
college.loc['University of Alaska Anchorage']
````

````python
# 重なりのない数集合を渡して行選択
college.iloc[[60,99,3]]
````

````python
# 指定したいindexのリストを作成して行選択
labels = ['University of Alaska Anchorage',
         'International Academy of Hair Design',
         'University of Alabama in Huntsville']
college.loc[labels]
````

````python
# スライス表記で選択
college.iloc[99:102]
````

````python
# locインデクサでスライス表記をすると、最後のラベルのデータも含まれる
start = 'International Academy of Hair Design'
stop = 'Mesa Community College'
college.loc[start:stop]
````

````python
# indexラベルの取り出し方
college.iloc[[60, 99, 3]].index
````

````python
# indexラベルを取り出してリスト化する
college.iloc[[60, 99, 3]].index.to_list()
````

## 28 DataFrameの行とカラムの同時選択

```
df.iloc[rows, columns]
df.loc[rows, columns]
```
rowsとcolomnsはスカラー値、リスト、スライスオブジェクト、Booleanのどれでも良い

````python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
````

````python
# 整数でスライス表記
college.iloc[:3, :4]
````

````python
# index名、column名でスライス表記
college.loc[:'Amridge University', :'MENONLY']
````

````python
# 異なるcolumnsの全行を選ぶ
college.iloc[:, [4,6]].head()
````

````python
# 異なるcolumnsの全行を選ぶ
college.loc[:,['WOMENONLY', 'SATVRMID']].head()
````
