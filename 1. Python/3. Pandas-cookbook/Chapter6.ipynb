## 49. 別のDataFrameからカラムを追加

```python
employee = pd.read_csv('Pandas-Cookbook-master/data/employee.csv')
dept_sal = employee[['DEPARTMENT', 'BASE_SALARY']]
```

```python
# リストの後ろから順番にソートされていく。最後にリストの最初の列でソートされる。
dept_sal = dept_sal.sort_values(['DEPARTMENT', 'BASE_SALARY'], ascending=[True, False])
dept_sal.head()
```

```python
# 各部門の先頭行を残す
max_dept_sal = dept_sal.drop_duplicates(subset='DEPARTMENT')
max_dept_sal.head()
```

```python
max_dept_sal = max_dept_sal.set_index('DEPARTMENT')
employee = employee.set_index('DEPARTMENT')
```

```python
employee['MAX_DEPT_SALARY'] = max_dept_sal['BASE_SALARY']
employee.head()
```

```python
employee.query('BASE_SALARY > MAX_DEPT_SALARY')
```

```python
# 非復元無作為抽出
np.random.seed(1234)
random_salary = dept_sal.sample(n=10).set_index('DEPARTMENT')
random_salary
```

```python
# 左のインデックスと１つは合致しないと結合できない
# 右辺の先頭3行と結合
employee['MAX_SALARY2'] = max_dept_sal['BASE_SALARY'].head(3)
employee.MAX_SALARY2.value_counts()
```

## 51. メソッドチェイニングでidxmaxの代用★★

```python
college = pd.read_csv('Pandas-Cookbook-master/data/college.csv', index_col='INSTNM')
college_desc = pd.read_csv('Pandas-Cookbook-master/data/descriptions/college_decsription.csv',index_col='Column Name')
```

```python
college_desc.loc[['MD_EARN_WNE_P10', 'GRAD_DEBT_MDN_SUPP']]
```

```python
cols = ['MD_EARN_WNE_P10', 'GRAD_DEBT_MDN_SUPP']
for col in cols:
    college[col] = pd.to_numeric(college[col], errors='coerce') # エラー : NaN
college_n = college.select_dtypes(include=[np.number]) # number列を選択したdf
criteria = college_n.nunique() == 2 # バイナリデータ列がtrue or false
binary_cols = college_n.columns[criteria].tolist() # バイナリデータがtrueの列リスト
college_n = college_n.drop(labels=binary_cols, axis='columns') # バイナリ列をdrop
college_n.head()
```

```python
# 各カラムの最大値を求める
college_n.max().head()
```

```python
# 各カラムの最大値かをテストする
# eq() : それぞれのカラムの最大値かテストする
college_n.eq(college_n.max()).head()
```

```python
has_row_max = college_n.eq(college_n.max()).any(axis='columns') # 各カラムで最大値を持つindexを取得
has_row_max.head()
```

```python
college_n.shape
```

```python
has_row_max.sum() # 最大値を持つindexの数
```

```python
college_n.eq(college_n.max()).cumsum() # cumsumで、各カラムごとの累積和を調べる
```

```python
college_n.eq(college_n.max()).cumsum().cumsum()
```

```python
has_row_max2 = college_n.eq(college_n.max()).cumsum().cumsum().eq(1).any(axis='columns')
has_row_max2
```

```python
has_row_max2.sum()
```

```python
idxmax_cols = has_row_max2[has_row_max2].index
idxmax_cols
```

```python
set(college_n.idxmax().unique()) == set(idxmax_cols)
```

## 52. 最多の最大値を求める

```python
college = pd.read_csv('Pandas-Cookbook-master/data/college.csv', index_col='INSTNM')
college_ugds = college.filter(like='UGDS_')
college.shape, college_ugds.shape
```

```python
highest_percentage_race = college_ugds.idxmax(axis='columns') # 各行で人種のパーセントが最大のカラムを取得する
highest_percentage_race.shape
```

```python
highest_percentage_race.head()
```

```python
# 最多人種の分布を確認
highest_percentage_race.value_counts(normalize=True)
```

```python
college_black = college_ugds[highest_percentage_race == 'UGDS_BLACK'] # 黒人男性が一番多い大学は？
college_black = college_black.drop('UGDS_BLACK', axis='columns') # 黒人割合カラムを削除
college_black.idxmax(axis='columns').value_counts(normalize=True) # 各行でmaxのカラム（黒人の次に大きいカラム）を取得→ユニーク名と出現回数を確認
```
