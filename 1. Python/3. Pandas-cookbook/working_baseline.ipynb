# 1. ライブラリ読み込み

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline 
```

# 2. データ確認 

```python
college = pd.read_csv('Pandas-Cookbook-master/data/college.csv')
college.head()
```

```python
# 次元確認
college.shape
```

```python
# データ要約、メモリ使用量
college.info()
```

```python
# 数値カラムの要約統計量を転置して確認
college.describe(include=[np.number]).T
```

```python
# オブジェクトカラムの要約統計量を転置して確認
college.describe(include=[np.object, pd.Categorical]).T
```

# 3. メモリ削減

```python
# 数値カラムを抽出
numcol = college.select_dtypes(include='number').columns.tolist()
```

```python
# 数値カラムの濃度を確認
college[numcol].nunique()
```

```python
# 数値カラムデータ型を確認
college[numcol].dtypes
```

```python
# 数値カラムのメモリ使用量を確認
college[numcol].memory_usage(deep=True)
```
