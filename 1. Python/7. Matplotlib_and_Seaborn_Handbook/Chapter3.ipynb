```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 3.1

```python
plt.plot([1, 2, 3, 4], [1, 4, 6, 16])
plt.ylabel("y-label")
plt.xlabel("x-label")
plt.show()
```

```python
# ラインを赤の点線にする
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 
         linestyle="dotted",
         linewidth="5",
         color="red")
plt.ylabel("y-label")
plt.xlabel("x-label")
plt.show()
```

```python
# 4種のラインを描画して凡例を表示する
x = np.array([1, 2, 3, 4])
y = np.array([1, 4, 9, 16])
# ラベル用テキストを関連付けてラインをプロット
plt.plot(x, y, linestyle="solid", label="Normal")
plt.plot(x, y/2, linestyle="dashed", label="Divided by 2")
plt.plot(x, y/3, linestyle="dashdot", label="Divided by 3")
plt.plot(x, y/4, linestyle="dotted", label="Divided by 4")
plt.legend()
plt.show()
```

## 3.2 グラフ操作のためのナビゲーションバー

## 3.3 軸のスケールを指定して散布図を描画する（pyplot.axis）

```python
# 散布図の書き方：linestyle=Noneにする
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], marker="o", linestyle="None")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# マーカーのスタイルを指定
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],
         marker="d",
         markersize=16,
         markerfacecolor="white",
         markeredgewidth=4,
         markeredgecolor="red",
         linestyle="None")
plt.axis([0, 5, 0, 20])
plt.show()
```

## 3.4 棒グラフの描画

```python
y = [15, 30, 45, 10, 5]
x = [1, 2, 3, 4, 5]
label= ["Apple", "Banana", "Orange", "Grape", "Strawberry"]
plt.bar(x=x,
        height=y,
        tick_label=label)
plt.title("Sales")
plt.xlabel("Fruit")
plt.ylabel("amout of sales")
plt.show()
```

```python
# バー間の隙間をなくす　バーの幅を指定するwidth
plt.bar(x=x,
        height=y,
        tick_label=label,
        width=1)
plt.show()
```

```python
# バーのカラーとエッジラインのスタイルを設定する
plt.bar(x=x,
        height=y,
        color="pink",
        edgecolor="red",
        linewidth=5)
plt.show()
```

```python
# 測定値の誤差（エラー）を示す
err = [i * 0.08 for i in y] # エラー値
plt.bar(x=x,
        height=y, 
        tick_label=label,
        yerr=err,
        ecolor="black",
        capsize=5)
plt.show()
```
