### 3.1.6 グラフを装飾する

```python
# 関数を定義
def f2(x,w):
    return (x-w) * x * (x+w)
```

```python
# xを定義
x = np.linspace(-3,3,100)
```

```python
# グラフ描写
plt.plot(x,f2(x,2),color='black',label='$w=2$')
plt.plot(x,f2(x,1),color='cornflowerblue',label='$w=1$')
plt.legend(loc='upper left')
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
import matplotlib
matplotlib.colors.cnames
```

```python
plt.figure(figsize=(10,3))
plt.subplots_adjust(wspace=0.5,hspace=0.5) # グラフの間隔を指定

for i in range(6):
    plt.subplot(2,3,i+1) # グラフ描写の位置を指定
    plt.title(i+1)
    plt.plot(x,f2(x,i),'k')
    plt.ylim(-20,20)
    plt.grid(True)
plt.show()
```

## 3.2 3次元のグラフを描く

### 3.2.1 2つの変数の関数

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 2変数関数f3(x0,x1)を定義
def f3(x0,x1):
    ans = (2 * x0**2 + x1**2) * np.exp(-(2 * x0**2 + x1**2))
    return ans
```

```python
# 各x0,x1でf3(x0,x1)を計算

# x0,x1は-2~2までの9つの数
xn = 9
x0 = np.linspace(-2,2,xn)
x1 = np.linspace(-2,2,xn)

# 9×9のゼロ行列を作成（初期化）
y = np.zeros((len(x0),len(x1)))
             
# (x0[i0],x1[i1])座標でf3を計算してyのi0行・i1列に代入
for i0 in range(xn):
    for i1 in range(xn):
        y[i1,i0] = f3(x0[i0],x1[i1])
```

```python
print(x0)
```

```python
print(np.round(y, 1))
```

### 3.2.2 数値を色で表現する：pcolor

```python
plt.figure(figsize=(3.5,3))
plt.bone() # 色調の指定
plt.pcolor(y) # 色による行列の表示
plt.colorbar() # カラーバーの表示
plt.show()
```

### 3.2.3 関数の表面を面で表す：surface

```python
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride=3, cstride=3, alpha=0.3, color='blue', edgecolor='black')
ax.set_zticks((0, 0.2))
ax.view_init(75, -95)
plt.show()
```

### 3.2.4 等高線で表示する：contour

```python
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((xn,xn))

for i0 in range(xn):
    for i1 in range(xn):
        y[i0,i1] = f3(x0[i0],x1[i1])
        
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4,4))
cont = plt.contour(xx0, xx1, y, 5, colors='black')
cont.clabel(fmt='%.2f',fontsize=8)
plt.xlabel('$x_0',fontsize=14)
plt.ylabel('$x_1',fontsize=14)
plt.show()
```
